```python
from pydrive.auth import GoogleAuth 
from pydrive.drive import GoogleDrive
import os
```

```python
gauth = GoogleAuth()
gauth.LocalWebserverAuth()
```

```python
drive = GoogleDrive(gauth)
```

```python
pipes_dir = 'pipes/posterior/'
file_list = os.listdir(pipes_dir)
```

```python
file = drive.CreateFile({'title': 'single_runs', 'parents' : [{'id' : '1Z1-ILMs-Z94gqODvfLlvLWWPA_mh4Xd_'}]})  

content_string = ''
for file_name in file_list:
    content_string += file_name + '\n'

file.SetContentString(content_string)

file.Upload()
```

```python
file_list = drive.ListFile({'q': "'1Z1-ILMs-Z94gqODvfLlvLWWPA_mh4Xd_' in parents and trashed=False"}).GetList()
for file in file_list:
  print('title: %s, id: %s' % (file['title'], file['id']))
```
